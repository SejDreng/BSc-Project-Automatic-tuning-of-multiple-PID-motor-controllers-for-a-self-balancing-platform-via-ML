```python
import numpy as np
import matplotlib.pyplot as plt

def plot_data_from_file(filename):
    # Read the data from the file, assuming it is space-separated
    data = np.loadtxt(filename)
    
    # Extract x and y values (second and third columns)
    x = data[:, 0]
    y = data[:, 4]
    # Create the plot
    plt.figure(figsize=(8, 6))
    plt.plot(x, y, label='Data', color='b')
    
    
    # Adding labels and title
    plt.xlabel('X values')
    plt.ylabel('Y values')
    plt.title('Plot from .dat file')
    
    # Show grid
    plt.grid(True)
    
    # Add a legend
    plt.legend()
    
    # Display the plot
    plt.show()
plot_data_from_file('ten_filters_filter.dat')
```
