```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib widget

data_ico = pd.read_csv('ICO_values.csv', index_col=False)
data_pid = pd.read_csv('pid_values.csv', index_col=False)
data_U = pd.read_csv('pos_U_values.csv', index_col=False)

headers_ico = data_ico.columns
headers_pid = data_pid.columns
headers_U = data_U.columns

# Create a new figure (ICO)
plt.figure( figsize=(12, 8.5), dpi=300)
for header in headers_ico:
    plt.plot(data_ico[header], label=header, linewidth=2)
plt.legend()
plt.grid(True)
plt.title('ICO Parameter Evolution')
plt.xlabel('Iteration')
plt.ylabel('ICO Values')

# Create a new figure (PID)
plt.figure(figsize=(12, 8.5))
for header in headers_pid:
    plt.plot(data_pid[header], label=header, linewidth=2)
plt.legend()
plt.grid(True)
plt.title('PID Parameter Evolution')
plt.xlabel('Iteration')
plt.ylabel('PID Values')

# dpi = 300
plt.figure( figsize=(12, 8.5))
i = 0;
for header in headers_U:
    i = i + 1;
    if i >= 3:
        break
    print(header)
    if header != 'U1' or header != 'U2':
        plt.plot(data_U[header], label=header, linewidth=2)
plt.legend()
plt.grid(True)
plt.title('Position evolution')
plt.xlabel('Iteration')
plt.ylabel('Radians')
```

```python
# Load data
data_ico = pd.read_csv('ICO_values.csv', index_col=False)
data_pid = pd.read_csv('pid_values.csv', index_col=False)
headers_ico = data_ico.columns
headers_pid = data_pid.columns

# Create a single figure with two subplots (1 row, 2 columns)
fig, axs = plt.subplots(2, 1, figsize=(12, 17), dpi=300)  # Adjust figsize as needed

# Plot ICO data on the first subplot
for header in headers_ico:
    axs[0].plot(data_ico[header], label=header, linewidth=2)
axs[0].legend()
axs[0].grid(True)
axs[0].set_title('ICO Parameter Evolution')
axs[0].set_xlabel('Iteration')
axs[0].set_ylabel('ICO Values')

# Plot PID data on the second subplot
for header in headers_pid:
    axs[1].plot(data_pid[header], label=header, linewidth=2)
axs[1].legend()
axs[1].grid(True)
axs[1].set_title('PID Parameter Evolution')
axs[1].set_xlabel('Iteration')
axs[1].set_ylabel('PID Values')

# Adjust layout to avoid overlap
plt.tight_layout()

# Display the plots
plt.show()
```
